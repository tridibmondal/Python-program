```python
cancer=pd.read_csv("E:\\datasets\\breast_cancer.csv")
```

```python
cancer
```

```python
cancer.head()
```

```python
cancer.groupby(by=['doctor_name','class']).size() #show all the doctors and class group 
```

```python
cancer.describe()
```

```python
cancer.shape

```

```python
# find the missing values
cancer.isnull() #gives all the values but in boolean that is true or false
cancer.isnull().sum()
```

```python
cancer.drop(['bland_chromatin'],axis=1 ,inplace=True)
```

```python
cancer.head()
```

```python
# fill the missing values
cancer['clump_thickness'].fillna(cancer['clump_thickness'].mean(),inplace=True)
```

```python
cancer.isnull().sum()
```

```python
cancer['bare_nuclei'].fillna(method="ffill",inplace=True)
```

```python
cancer.isnull().sum()
```

```python
#converting string to numerical value 1. get_dummies() packagr name pandas 2. label encoder() package name sklearn

newclass=pd.get_dummies(cancer['class'])
```

```python
newclass
```

```python
# to drop class column
cancer.drop(['class'],axis=1,inplace=True)
```

```python
cancer.head()
```

```python
# to add new class to cancer data frame
cancer=pd.concat([cancer,newclass],axis=1)
```

```python
cancer.head()
```
