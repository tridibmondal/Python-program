```python
import sqlite3
mydb=sqlite3.connect("Sample2.db")
c=mydb.cursor()
print("----- BANK MANAGEMENT SYSTEM -----")

while True:
    ch=int(input("""1.Admin\n2.Customer\n3.Exit\nEnter choice :"""))
    if(ch==1):
        print("\n -- Welcome ADMIN -- \n")
        
        while True:
            
            v=int(input("1.View Customer\n2.Create Account\n3.exit\nEnter your Choice :"))
        
            if(v==1):
                print("\n -- Customer Details : -- \n")
                c.execute("select * from Bank")
                res=c.fetchall()
                for i in res:
                    print(i)
        
            elif(v==2):
                print("\n -- Creating New Account : -- \n")
                uid=input("Enter the Id :")
                uname=input("Enter the Name :")
                ubalance=input("Enter the Balance")
                c.execute("insert into Bank values(?,?,?)",(uid,uname,ubalance))
                mydb.commit()
        
            elif(v==3):
                break
    
    
    elif(ch==2):
        print("\n -- Welcome CUSTOMER -- \n")
        
        uid=input("Enter your id :")
        c.execute('select name from Bank where id=?',(uid,))
        res1=c.fetchall()
            
        if(res1):
            n=res1[0][0]
            print("\n -- Welcome to your account ,",n,' -- \n')
            while True:
                    
                    ch1=int(input("1.Checkbalance\n2.Deposit\n3.widthdraw\n4.Exit\nEnter your Choice :"))
                    
                    if(ch1==1):
                        print("\n -- Displaying Balance : -- \n")
                        c.execute('select balance from Bank where id=?',(uid,))
                        bal=c.fetchall()
                        x=int(bal[0][0])
                        print("Your Balance is :",x)
                        
                    elif(ch1==2):
                        print("\n -- Cash Deposition : -- \n")
                        c.execute('select balance from Bank where id=?',(uid,))
                        bal=c.fetchall()
                        x=int(bal[0][0])
                        D=int(input("enter the amount to be deposited :"))
                        D=D+x
                        c.execute('update Bank set balance=? where id=?',(D,uid,))
                        print("balance updated")
                        mydb.commit()
                        
                    elif(ch1==3):
                        print("\n -- Cash Withdrawl : -- \n")
                        c.execute('select balance from Bank where id=?',(uid,))
                        bal=c.fetchall()
                        x=int(bal[0][0])
                        W=int(input("enter the amount to be Widthdraw :"))
                        if(x<=W):
                            print("Insufficent balance amount cannot be withdrawn :")
                        else:
                            W=x-W
                            c.execute('update Bank set balance=? where id=?',(W,uid,))
                            print("balance updated")
                            mydb.commit()
                            
                    elif(ch1==4):
                        break
        else:
                print(" Sorry !!! ID not found .. ")
    elif(ch==3):
        break

                                 
```
